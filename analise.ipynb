```python
import pandas as pd
def gerar_tabelas_frequencia(df, coluna):
    if isinstance(coluna, int):
        coluna_nome = df.columns[coluna]
    else:
        coluna_nome = coluna

    if df[coluna_nome].dtype == 'O':
        # Tabela de frequências absolutas
        frequencia_absoluta = df[coluna_nome].value_counts()

        # Tabela de frequências relativas
        frequencia_relativa = df[coluna_nome].value_counts(normalize=True)

        # Criando um DataFrame com as duas frequências
        tabela_completa = pd.DataFrame({'Frequência Absoluta': frequencia_absoluta, 'Frequência Relativa': frequencia_relativa})

        return tabela_completa
    else:
        return None
```

## Probalidade e associação

```python
import pandas as pd
from scipy.stats import chi2_contingency

variable_resposta = 'Boa_Performance'


for column in dados.columns:
    if column != variable_resposta and column != 'Id_Agente':
        contingency_table = pd.crosstab(dados[column], dados[variable_resposta])
        chi2, p, _, _ = chi2_contingency(contingency_table)
        
        print(f"Coluna: {column}")
        print(f"Chi-squared: {chi2}")
        print(f"P-value: {p}")
        
        if p < 0.05:
            print("Há uma associação significativa.\n")
        else:
            print("Não há uma associação significativa.\n")
```
